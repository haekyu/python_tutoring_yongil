### Loop 문
- 뭔가 반복할 일이 생길 때 사용
- while
    - 만드는 방법
        ```
        while 조건:
            do
        ```
    - 예시
        ```
		n = 0
		while n < 10:
			print(n)
			n = n + 1
        ```
- for
    - 만드는 방법
        ```
        for iterator in sequence:
			do
        ```
    - 자주 쓰는 유형
        + range와 함께 사용
            + 기본
                ```
                for i in range(from이상, to미만, stepsize):
                    do
                ```
            + range의 input이 두 개일 때
                - stepsize = 1 (default)
                ```
                for i in range(from이상, to미만):
                    do
                ```
            + range의 input이 한 개일 때
                - stepsize = 1 (default)
                - from = 0 (default)
                ```
                for i in range(to미만):
                    do
                ```
    - enumerate
        - sequence와 sequence의 차례를 동시에 알고싶을 때
        - 아래 두 개를 
            ```
            for e in lst:
                ssdf
            for idx in range(len(lst)):
                sdfsdf
            ```
        - 이렇게 합친 것과 같음
            ```
            for idx, e in enumerate(lst):
                sfsdfsdfs
            ```
    - zip
        - sequence를 묶어서 한꺼번에 이터레이션 돌고 싶을 때
        - sequence 들의 길이가 같아야 함
        - 사용 규칙
            ```
            for i1, i2, ..... in zip(s1, s2, ....):
                sdfsdfds
            ```
        - 예)
            ```
            name = ['철수', '영희', '영수']
            age = [1, 2, 3]
            place = ['aaa', 'bbb', 'ccc']

            for n, a, p in zip(name, age, place):
                print('이름=', n, '나이=', a, '거주지=', p)
            ```
		
- Loop 문에 변칙주기
    - break
        = 중도 멈춤
    - continue
        = skip
	

````python
i = 0
while i < 10:
    print(i)
    i = i + 1
````

````python
# n = 1 + 2 + 3 + .... + 10 == 55
n = 0
for i in range(1, 10 + 1):
    n += i
print(n)
````

### 함수
- 만드는 규칙
    ```
	def 함수이름(인풋1, 인풋2, 인풋3, ...):
		어쩌고 저쩌고
		return 아웃풋1, 아웃풋2, ....
	```
- 예
    ```
	def g(x, y):
	    a = x + y
	    b = x * y
	    c = x - y
	    return a, b, c
    ```

````python
def g(x, y):
    a = x + y
    b = x * y
    c = x - y
    return a, b, c


result = g(3, 3)
print(result)
print(result[1])

r1, r2, r3 = g(3, 3)
print(r2)
````

### str fmt
- string 만드는 규칙
- '안에다가 스트링 패턴 만들고' % (패턴 내에 대입할 값을 튜플로 만듦)
    - 예)
        ```
        >>> 'mmmmmm   %d sssss %r' % (1, True)
        'mmmmmm   1 sssss True'
        ```
	- 예2)
        ```
		print('idx=', idx, 'e=', lst[idx])
		print('idx=%d, e=%lf' % (idx, lst[idx]))
        ```
- 패턴 규칙
	- %d == int
	- %lf == float
	- %s == string
	- %r == bool
